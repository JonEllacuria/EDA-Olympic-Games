```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objs as go
from plotly.subplots import make_subplots
from plotly.offline import init_notebook_mode, iplot, plot
import plotly.offline as pyo
import matplotlib
```

```python
repetidos_total_2=pd.read_csv("repetidos_total_2.csv")
```

```python
media_edad=repetidos_total_2.groupby("Year")["Age"].mean()
```

```python
media_edad
```

```python
media_edad=repetidos_total_2.groupby(["Year","Sex"])["Age"].mean()
```

```python
media_edad
```

```python
media_edad_desagregada = media_edad.unstack()

# Creamos una figura con un solo eje
fig, ax = plt.subplots()

# Trazamos la línea para hombres en azul
media_edad_desagregada["M"].plot(kind='line', ax=ax, color='green', label='Male')

# Trazamos la línea para mujeres en rojo
media_edad_desagregada["F"].plot(kind='line', ax=ax, color='red', label='Female')

# Agregamos etiquetas y títulos
ax.set_xlabel('Years')
ax.set_ylabel('Age')
ax.set_title('Mean Age')
ax.legend()
ax.grid(False)
sns.set_style("whitegrid")
plt.savefig("Mean Age_Total", transparent=True)

plt.show()
```

```python
medallistas=pd.read_csv("medallistas.csv")
```

```python
medallistas=medallistas[medallistas["Sex"].isin(["M","F"])]
```

```python
medallistas_media_edad=medallistas.groupby(["Year","Sex"])["Age"].mean()
```

```python
medallistas_media_edad
```

```python
medallistas_media_edad_desagregada = medallistas_media_edad.unstack()

# Creamos una figura con un solo eje
fig, ax = plt.subplots()

# Trazamos la línea para hombres en azul
medallistas_media_edad_desagregada["M"].plot(kind='line', ax=ax, color='green', label='Male')

# Trazamos la línea para mujeres en rojo
medallistas_media_edad_desagregada["F"].plot(kind='line', ax=ax, color='red', label='Female')

# Agregamos etiquetas y títulos
ax.set_xlabel('Years')
ax.set_ylabel('Age')
ax.set_title('Mean Age Medalists')
ax.legend()
ax.grid(False)
sns.set_style("whitegrid")
plt.savefig("Mean Age_Medallistas", transparent=True)

plt.show()
```
