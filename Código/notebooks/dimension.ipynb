```python
unicos_total_2 = repetidos_total_2.drop_duplicates(subset=['Year', 'Name'])

```

```python
# Agrupar los datos por año y género, contar el número de participantes de cada grupo
gender_comparation = unicos_total_2.groupby(['Year', 'Sex'])['Name'].count().reset_index()

# Reorganizar los datos para que cada género sea una columna y cada fila represente el año
gender_comparation_plot = gender_comparation.pivot(index='Year', columns='Sex', values='Name')

# Crear el gráfico de barras apiladas
gender_comparation_plot.plot(kind='bar', stacked=True,color=["green","orange"])

# Establecer las etiquetas de los ejes y el título
plt.xlabel('Year')
plt.ylabel('Athletes')
plt.title('Total Athletes')
ax = gender_comparation_plot.plot(kind='bar', stacked=True, color=["green","orange"])

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[:2], labels[:2], loc='best')

sns.set_style("whitegrid")
plt.savefig("Total Athletes", transparent=True)

# Mostrar el gráfico
plt.show()
```

```python
medallistas=repetidos_total_2[repetidos_total_2["Medal"].notna()]
```

```python
medallistas["Medal"].unique()
```

```python
medallistas['Medal'] = medallistas['Medal'].replace('Gold Medal', 'Gold')
medallistas['Medal'] = medallistas['Medal'].replace('Silver Medal', 'Silver')
medallistas['Medal'] = medallistas['Medal'].replace('Bronze Medal', 'Bronze')
```

```python
medallistas_agrupados = medallistas.groupby(['Year', 'Medal'])['Name'].count().reset_index()
paleta_medallas = ['#CD7F32','#D4AF37','#add8e6']
orden_medallas=["Bronze","Silver","Gold"]
# Hacer el catplot utilizando los datos agregados
sns.catplot(x="Year",
            y="Name",
            hue="Medal",
            kind="bar",
            data=medallistas_agrupados,
            palette=paleta_medallas,
            #order=orden_medallas,
            aspect=1.5,
            ci=None,);
plt.xticks(rotation=45)
plt.yticks([]) # elimina los valores del eje y
plt.ylabel(" ")

sns.set_style("whitegrid")

plt.savefig("Number of Medals_2", transparent=True)
```
